# **1.) Apriori algorithm implemented with the apyori package for Market Basket Optimization.**

```python
#importing required packages
import pandas as pd
import numpy as np
!pip install apyori
from apyori import apriori
```

```python
#loading market basket dataset..

df = pd.read_csv('data-1.csv',header=None)
df.head()
```

```python
# replacing empty value with 0. (Data cleaning)
df.fillna(0,inplace=True)
df.head()
```

```python
# Data Pre-processing
# Converting data in list format to for using aprori

transactions = []
for i in range(0,len(df)):
    transactions.append([str(df.values[i,j]) for j in range(0,20) if str(df.values[i,j])!='0'])
transactions[0:3]
```

```python
# Calling apriori function
rules = apriori(transactions, min_support=0.003, min_confidance=0.2, min_lift=3, min_length=2)
rules
Results = list(rules)
## min_support = 0.003 -> means selecting items with min support of 0.3%
## min_confidance = 0.2 -> means min confidance of 20% 
## min_lift = 3  
## min_length = 2 -> means no. of items in the transaction should be 2
```

```python
# creating dataframe of the rules
df_results = pd.DataFrame(Results)
support = df_results.support
df_results.head()
```

```python
#Creating 4 list for optimization
first_values = []
second_values = []
third_values = []
fourth_value = []

for i in range(df_results.shape[0]):
    single_list = df_results['ordered_statistics'][i][0]
    first_values.append(list(single_list[0]))
    second_values.append(list(single_list[1]))
    third_values.append(single_list[2])
    fourth_value.append(single_list[3])
```

```python
# creating data frame of all lists
lhs = pd.DataFrame(first_values)
rhs = pd.DataFrame(second_values)
confidance=pd.DataFrame(third_values,columns=['Confidance'])
lift=pd.DataFrame(fourth_value,columns=['lift'])
```

```python
# concat all list together in a single dataframe
df_final = pd.concat([lhs,rhs,support,confidance,lift], axis=1)
df_final.head()
```

```python
df_final.fillna(value=' ', inplace=True)
df_final.head()
```

```python
#set column name
df_final.columns = ['lhs',1,'rhs',2,3,'support','confidance','lift']
df_final.head()   
```

```python
# Adding column 1 in lhs and 2,3 in rhs then dropping them
df_final['lhs'] = df_final['lhs'] + str(", ") + df_final[1]
df_final['rhs'] = df_final['rhs']+str(", ")+df_final[2] + str(", ") + df_final[3]
df_final.drop(columns=[1,2,3],inplace=True)
df_final.head()
```

```python
#Displaying the top 5 values after sorting by lift
df_final.sort_values('lift', ascending=False).head()
```

# **2.) Apriori algorithm implemented for Market Basket Optimization using the mlxtend.frequent_patterns package.**

Why are we using this method?

1. The first approach had a limitation of requiring data conversion into a list format, which can be computationally expensive for stores with a large number of transactions.

2. The Apyori package is outdated and may not offer the latest functionalities.

3. The results obtained from the previous approach are presented in a less user-friendly format, indicating the need for additional pre-processing steps.

4. Instead, we can utilize the mlxtend package, which not only generates frequent itemsets but also provides association rules, addressing the limitations mentioned earlier.

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

df1 = pd.read_csv('data-2.csv', encoding="ISO-8859-1")
df1.head()
```

```python
df1.Country.value_counts()
```

```python
df1 = df1[df1.Country == 'France'] # Since data has many country choosing any one for check.
```

```python
df1['Description'] = df1['Description'].str.strip() # Remoiving spaces
df1 = df1[df1.Quantity >0] # removing transcation with neagtive values
df1.head()
```

```python
# creating pivot table with Quantity sum as values and filling 0 if any nan values

basket = pd.pivot_table(data=df1,index='InvoiceNo',columns='Description',values='Quantity', aggfunc='sum',fill_value=0)
basket.head()
```

```python
# checking correctness after binning to 1 
basket['10 COLOUR SPACEBOY PEN'].head(6)
```

```python
# converting to 1 if more then 0 else o
def convert_into_binary(x):
    if x > 0:
        return 1
    else:
        return 0
basket_sets = basket.applymap(convert_into_binary)
basket_sets['10 COLOUR SPACEBOY PEN'].head(6)
```

```python
# removing postage since its insignificant
basket_sets.drop(columns=['POSTAGE'],inplace=True)
```

```python
# calling apriori function and passing minimum support.
frequent_itemsets = apriori(basket_sets, min_support=0.07, use_colnames=True)
frequent_itemsets.head()
```

```python
# setting metric as lift and keeping minimum lift as 1
rules_mlxtend = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules_mlxtend.head()
```

```python
# Displaying items with lift >= 4 and confidence >= 0.8
rules_mlxtend[ (rules_mlxtend['lift'] >= 4) & (rules_mlxtend['confidence'] >= 0.8) ]
```

1. Antecedent support: This measure indicates the frequency of the antecedent item in all the transactions. For example, the item "ALARM CLOCK BAKELIKE RED" is present in 9.4% of the transactions.

2. Consequent support: This measure indicates the frequency of the consequent item in all the transactions. For example, the item "ALARM CLOCK BAKELIKE GREEN" is present in 9.6% of the transactions.

3. Support: This measure indicates the frequency of the itemset in all the transactions. For example, the itemset {ALARM CLOCK BAKELIKE RED, ALARM CLOCK BAKELIKE GREEN} is present in 7.9% of the transactions.

4. Confidence: This measure defines the likelihood of finding the consequent item in the cart given that the cart already contains the antecedents. For example, {ALARM CLOCK BAKELIKE RED} -> {ALARM CLOCK BAKELIKE GREEN} has a confidence of 83%. In simple terms, there is an 83% chance of finding {ALARM CLOCK BAKELIKE GREEN} if the cart contains {ALARM CLOCK BAKELIKE RED}.

5. Lift: This measure defines the likelihood of finding the consequent item in the cart given that the cart already contains the antecedent, while considering the popularity of the consequent item. For example, the lift of {ALARM CLOCK BAKELIKE GREEN} with respect to {ALARM CLOCK BAKELIKE RED} is 8.64, indicating a strong association. A lift value greater than 1 implies that the association rule is worth considering.

6. Leverage: Leverage measures the difference between the occurrence of X and Y appearing together in the dataset and what would be expected if X and Y were statistically independent. In a sales setting, it helps determine the additional units (items X and Y together) sold compared to what would be expected from independent sales.
